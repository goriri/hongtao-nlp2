```python
import pandas as pd

df1 = pd.read_csv('Rock-data-Aug.csv',encoding='utf-8')
df1
```

```python
from random import shuffle
import jieba
import re
```

```python
file  = 'dataset.csv'#'toutiao_cat_data.txt'
with open(file) as f:
    lines = f.readlines()
    
labels = []
#i=0
for line in lines:
    label = []
    line = line.split(',')
    label_code = line[0]
    #print(i)
    #i += 1
    label.append(label_code)
    line[1] = re.sub(r"[\s+\.\!\/_,$%^*()?;；:-【】+\"\']+|[+——一！，;:：。？、~@#￥%……&*（）]+", "", line[1])
    #print(line[1])
    label.extend(jieba.cut(line[1],cut_all=False))
#     print(label)
#     label = '__label__' + label_code + ' ' + artical_title
    labels.append(label)
    
shuffle(labels)
print(labels[0:5])
```
